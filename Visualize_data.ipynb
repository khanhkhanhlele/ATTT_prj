#VirusShare

```python
VirusShare = pd.read_csv("api_sequences_malware_datasets/VirusShare.csv")
df = pd.DataFrame(VirusShare)
df.head()
```

```python
sns_plot = sns.countplot(data=VirusShare,y ='class')
fig = sns_plot.get_figure()
fig.savefig('VirusShare_hist.png')
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import stopwords
stop=set(stopwords.words('english'))
def preprocess_VirusShare(df):
    corpus=[]
    stem=PorterStemmer()
    lem=WordNetLemmatizer()
    for virus in df['api']:
        words=[w for w in word_tokenize(virus) if (w not in stop)]

        words=[lem.lemmatize(w) for w in words if len(w)>2]

        corpus.append(words)
    return corpus

corpus=preprocess_VirusShare(VirusShare)
```

```python
from wordcloud import WordCloud, STOPWORDS
from matplotlib import figure
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
stopwords = set(STOPWORDS)

def show_wordcloud(data):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=stopwords,
        max_words=100,
        max_font_size=30,
        scale=3,
        random_state=1)

    wordcloud=wordcloud.generate(str(data))

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')

    plt.imshow(wordcloud)
    fig.set_canvas(plt.gcf().canvas)
    fig.savefig('VirusShare_wordcloud.png')
    plt.show()
```

###VirusShare_wordcloud

```python
show_wordcloud(corpus)
```

###VirusShare_wordcloud_Virus

```python
df_new = VirusShare[VirusShare['class'] == 'Virus']
corpus=preprocess_VirusShare(df_new)
show_wordcloud(corpus)
```

###VirusShare_wordcloud_Adware

```python
df_new = VirusShare[VirusShare['class'] == 'Adware']
corpus=preprocess_VirusShare(df_new)
show_wordcloud(corpus)
```

###VirusShare_wordcloud_Undefined

```python
df_new = VirusShare[VirusShare['class'] == 'Undefined']
corpus=preprocess_VirusShare(df_new)
show_wordcloud(corpus)
```

###VirusShare_wordcloud_Downloader

```python
df_new = VirusShare[VirusShare['class'] == 'Downloader']
corpus=preprocess_VirusShare(df_new)
show_wordcloud(corpus)
```
